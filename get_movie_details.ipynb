```python
def fetch_movie_credits_by_id():
    url = 'https://api.themoviedb.org/3/movie/' 
    api_key = '/credits?api_key=308b04f2b1399d2700c3d6209623e241&language=en-US'
    data_file_cast = open(curr_dir + 'cast_details.csv', 'w',encoding='utf-8-sig')
    data_file_crew = open(curr_dir + 'crew_details.csv', 'w',encoding='utf-8-sig')
    csv_writer_cast = csv.writer(data_file_cast)
    csv_writer_crew = csv.writer(data_file_crew)
    write_keys_cast = True
    write_keys_crew = True
    with open(curr_dir + 'Movie_list.csv','r',encoding='utf-8-sig') as csvfile:
        f = csv.reader(csvfile,delimiter=',',quotechar = '"')
        for line in f:
            if line[0] == 'adult' and line[1] == 'backdrop_path':
                next
            else:
                mainURL = url + str(line[3]) + api_key
                try:
                  response = requests.get(mainURL)
                  response_json = response.json()
                except:
                  continue
                
                
                #if write_keys:
                #    csv_writer_cast.writerow(results_cast.keys())
                #    csv_writer_crew.writerow(results_crew.keys())
                #    write_keys = False
                try:
                  results_cast = response_json['cast']
                  for i in range(5):
                      try:
                          result = results_cast[i]
                          result['movie_id'] = response_json['id']
                          if write_keys_cast:
                              csv_writer_cast.writerow(result.keys())
                              write_keys_cast = False
                          csv_writer_cast.writerow(result.values())
                      except:
                          print(str(i) + " cast not found")
                except:
                  print("no cast found") 

                try:  
                  results_crew = response_json['crew']   
                  for result in results_crew:
                      if result['job'] == 'Director':
                          result['movie_id'] = response_json['id']
                          if write_keys_cast:
                              csv_writer_cast.writerow(result.keys())
                              write_keys_cast = False
                          csv_writer_crew.writerow(result.values())
                except:
                  print("no crew found")
                

    data_file_cast.close()
    data_file_crew.close()
```

```python
def fetch_collection_details_by_id():
    url = 'https://api.themoviedb.org/3/collection/' 
    api_key = '?api_key=308b04f2b1399d2700c3d6209623e241&language=en-US'
    data_file = open(curr_dir + '\Collection_details.csv', 'w',encoding='utf-8-sig')
    csv_writer = csv.writer(data_file)
    write_keys = True
    with open(curr_dir + 'Movie_details.csv','r',encoding='utf-8-sig') as csvfile:
        f = csv.reader(csvfile,delimiter=',',quotechar = '"')
        for line in f:
            if len(line) == 0:
                next
            else:
                if (line[0] == 'adult' and line[1] == 'backdrop_path'):
                    next
                else:
                  try:
                    if line[2] != '':

                        mainURL = url + str(line[2].split(',')[0].split(':')[1]) + api_key
                        response = requests.get(mainURL)
                        response_json = response.json()
                        results = response_json

                        if write_keys:
                            csv_writer.writerow(results.keys())
                            write_keys = False

                        #for result in results:
                        csv_writer.writerow(results.values())
                  except:
                    print(line)

    data_file.close()
```
